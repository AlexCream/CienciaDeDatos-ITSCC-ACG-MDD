# Actividad 4: Agrupaciones Multinivel con Pandas

Con Pandas no solo podemos hacer agrupaciones simples, podemos agrupar por **múltiples columnas al mismo tiempo** (multinivel).

**Métodos clave:**
- `groupby(["col1", "col2"])` — agrupa por dos o más columnas
- `get_group((valor1, valor2))` — extrae un grupo específico del multinivel
- `.agg()` / `.mean()` — calcula estadísticas sobre los grupos

## PASO 0: Cargar el dataset e importar dependencias

```python
import pandas as pd
import datetime

df = pd.read_csv("https://raw.githubusercontent.com/AlexCream/MultinivelDeDatos/main/pipol_dataset.csv")

# Convertir birthday a edad
df["birthday"] = pd.to_datetime(df["birthday"], dayfirst=True)
df["age"] = datetime.datetime.now().year - df["birthday"].dt.year

print("Dataset con columna age:")
print(df[["name", "city", "country", "age"]].head(10))
```

## Parte a) Agrupacion multinivel: por pais Y ciudad

Cuando agrupamos por **dos columnas**, Pandas crea un indice jerarquico.
Para obtener un grupo especifico usamos `get_group()` con una **tupla**: `(pais, ciudad)`.

```python
# Agrupacion multinivel: pais + ciudad
multi_group = df.groupby(["country", "city"])

print(f"Total de combinaciones pais-ciudad: {multi_group.ngroups}")
print()
print("Grupos disponibles (pais, ciudad):")
print(*(list(multi_group.groups.keys())), sep="\n")
```

```python
# Extraer el grupo especifico: Mexico y Dusty City
# get_group recibe una TUPLA cuando el groupby es multinivel
mexico_dusty = multi_group.get_group(("Mexico", "Dusty City"))

print("Personas en Mexico - Dusty City")
print(mexico_dusty[["name", "last_name", "city", "country", "age"]].to_string(index=False))
```

## Parte b) Edad promedio por pais

Podemos calcular estadisticas a nivel de pais agrupando solo por `country`.
Tanto `.mean()` como `.agg()` nos dan el mismo resultado; la diferencia es que `.agg()` es mas flexible.

```python
# Metodo 1: .mean()
group_pais = df.groupby("country")

print("Edad promedio por pais con .mean():")
print(group_pais["age"].mean().round(2))
```

```python
# Metodo 2: .agg("mean") — equivalente, pero permite pasar multiples funciones
print("Edad promedio por pais con .agg():")
print(group_pais["age"].agg("mean").round(2))
```

```python
# BONUS: edad promedio por pais Y ciudad usando el grupo multinivel
print("Edad promedio por pais y ciudad:")
print(multi_group["age"].mean().round(2))
```
