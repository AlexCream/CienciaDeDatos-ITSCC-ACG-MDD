```python
# Crear código para un DataFrame para poder leer el archivo utilizando Pandas y el comando "path" para acceder desde Github 
# Debe utilizar un comando para leer el tipo de archivo, index_col=[0] para usar la primera columna como índice.
import pandas as pd
from scipy import stats
import numpy as np
path = "https://raw.githubusercontent.com/AlexCream/CienciaDeDatos-ITSCC-ACG-MDD/main/pipol_datos.csv"
data = pd.read_csv(path,index_col=[0])
data.columns = ["age", "sex", "income","height", "city", "education", "children"]
print(data.info())
```

```python
data_edad = data["age"]
print("Validacion del campo edad")
print("Nulos:",data_edad.isnull().sum())
print("Negativos:",data_edad[data_edad<0].shape[0])
print("Valor mínimo:",data_edad.min())
print("Valor máximo:",data_edad.max())
```

```python
data_edad_nonegativo = data_edad.copy()
data_edad_nonegativo[data_edad_nonegativo < 0] = 0
print("Negativos:",data_edad_nonegativo[data_edad_nonegativo<0].shape[0])
print("Valor mínimo:",data_edad_nonegativo.min())
print("Valor máximo:",data_edad_nonegativo.max())
```

```python
for inicio in range(0, 200, 10):
    fin = inicio + 10
    operador = "<=" if fin == 200 else "<"
    count = data_edad_nonegativo[(data_edad_nonegativo >= inicio) & (data_edad_nonegativo < fin)].shape[0]
    print(f"Valores entre {inicio}-{fin}: {count}")
```

```python
z_scores = np.abs(stats.zscore(data_edad_nonegativo))
data_edad_zscore = data_edad_nonegativo[z_scores < 2]
for inicio in range(0, 200, 10):
    fin = inicio + 10
    operador = "<=" if fin == 200 else "<"
    count = data_edad_zscore[(data_edad_zscore >= inicio) & (data_edad_zscore < fin)].shape[0]
    print(f"Valores entre {inicio}-{fin}: {count}")
```

```python
z_scores = np.abs(stats.zscore(data_edad_nonegativo))
limite_superior = data_edad_nonegativo[z_scores < 2].max()
data_edad_zscore_trunco = data_edad_nonegativo.copy()
data_edad_zscore_trunco[z_scores >= 2] = limite_superior

for inicio in range(0, 200, 10):
    fin = inicio + 10
    if fin == 200:
        count = data_edad_zscore_trunco[(data_edad_zscore_trunco >= inicio) & (data_edad_zscore_trunco <= fin)].shape[0]
    else:
        count = data_edad_zscore_trunco[(data_edad_zscore_trunco >= inicio) & (data_edad_zscore_trunco < fin)].shape[0]
    print(f"Valores entre {inicio}-{fin}: {count}")
```
