##### Declaracion de una funcion lambda, en la cual se evaluan dos condiciones,
##### 1. Que ambas cadenas tengan los mismos caracteres (por eso se usa sorted para que ambas cadenas esten ordenadas igual)
##### 2. Que la cadena dos pueda encontrarse como subconjunto de la concatenacion entre l. cadena uno y la cadena uno,

```python
are_rotations = lambda s_1, s_2: sorted(s_1) == sorted(s_2) and s_2 in s_1 + s_1
```

##### Se imprime el resultado de la funcion are_rotations, cuando como parametros o entrada se le ingresa las cadenas abcde y cdeab, las cuales al someterse a las reglas y condiciones nos arroja true

```python
print(are_rotations("abcde","cdeab"))
```

##### Se imprime el resultado de la funcion are_rotations, cuando como parametros o entrada se le ingresa las cadenas abcde y abced, las cuales al someterse a las reglas y condiciones nos arroja false

```python
print(are_rotations("abcde","abced"))
```
