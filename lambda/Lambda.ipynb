```python
# Repasando el uso de Lambda, muestre ejemplos prácticos en Python. 
## Elimine de la cadena definida "Python es super divertido" las vocales y guarde y muéstrelos en otra cadena .
### Creacion de la cadena base
cadena = "Python es super divertido"
### Creacion del conjunto de caracteres que se quieren eliminar (grupo de vocales)
vocales = "aeiouAEIOU"
```

```python
### Ejecucion de la funcion lambda que se encarga de buscar los elementos de nuestra cadena en 
### el conjunto y que en caso de encontrarlos los borrara de la cadena original
cadena_sin_vocales = ''.join(filter(lambda x: x not in vocales, cadena))
```

```python
### Impresion de la cadena generada sin vocales
print(cadena_sin_vocales)
```

```python
## Contar los caracteres de la siguiente cadena definida "Hello World Hola Mundo".
cadena = "Hello World Hola Mundo"
```

```python
### Crear la funcion contador de caracteres, que remplace los espacios vacios por nada
contador_caracteres = lambda s: len(s.replace(" ", ""))
```

```python
### Imprimir el resultado de la cadena
print ("La cadena ",cadena," tiene ",contador_caracteres(cadena)," caracteres que no son espacios en blanco")
```

```python
# Validar si una cadena es la traslación de otra, utilice la cadena definida "abcdef".
## Generamos la cadena base que nos indica la practica
cadena1 = "abcdef"
## Generamos una cadena que se trata de una traslacion de la cadena base
cadena2 = "bcdefa"
## Generamos una cadena completamente diferente
cadena3 = "salamalecu"
```

```python
## Creamos la funcion lambda que verifica primero el tamaño de la cadena y 
## posteriormente busca que la cadena aparesca dentro de la combinacion (suma) de la cadena base
es_translacion = lambda s1, s2: len(s1) == len(s2) and s2 in s1 + s1
```

```python
## Llamamos a la funcion lambda para que compare las cadenas y por medio de un if nos indique si las 
## cadenas son/no son traslacion unas de las otras
print(cadena1, "y", cadena2, "son" if es_translacion(cadena1, cadena2) else "no son", "traslación")
print(cadena1, "y", cadena3, "son" if es_translacion(cadena1, cadena3) else "no son", "traslación")
```
