```python
# 1) La función de np.random: Colocar 2 ejemplos en código. 
import numpy as np
rng = np.random.default_rng()

# Cara o cruz
while True:
    X = input("Tirar una moeda al azar? [S/N]: ")
    if X.lower() != "s":
        break
    if rng.random()>0.5:
        print("Cara")
    else:
        print("Cruz")
    input("Presione enter para continuar")
```

```python
# Arreglo lleno con numeros random
print("Creacion de una matriz nxn de numeros random")
while True:
    X = input("Igresa la cantidad de columnas")
    if X.isdigit():
        break
    else:
        print("Ingresa un valor entero")
while True:
    Y = input("Igresa la cantidad de renglones")
    if Y.isdigit():
        break
    else:
        print("Ingresa un valor entero")
X = int(X)
Y = int(Y)
print("Matriz de enteros [1-100]")
print(np.random.randint(0, 100, (Y, X)) )
print("Matriz de decimales [0-1]")
print(np.random.rand(Y, X) )
```

```python
# 2) La función de np.linspace: Colocar 2 ejemplos en código.
a = np.linspace(0, 1, num=9)
print(a)
```

```python
#retstep=True que tambien retorne de cuando es el avanze entre cada numero
#num=5 especifica que se generen 5 valores separados de forma equivalente
array, c = np.linspace(0, 100, num=5, retstep=True)
print("Paso:", c)
print("Arreglo:", array)
```

```python
# 3) Operaciones Matemáticas Básicas: Se escogerán las que no se han utilizado (Exponenciación, Raíz cuadrada, Logaritmos en varias bases).
# endpoint=False excliye e primer valor del resultado.
arreglobase = (np.linspace(0, 10, num=10, endpoint=False).reshape(2, 5))+1
print("Arreglo inicial")
print(arreglobase)
```

```python
# Raices de todo el arreglo
print(np.sqrt(arreglobase))
```

```python
# Exponenciacion
print("Cuadrado")
print(np.power(arreglobase,2))

print("Cubo")
print(np.power(arreglobase,3))

print("Exponenciacion de euler")
print(np.exp(arreglobase))
```

```python
# Logaritmo
print("Logaritmo en base 2")
print(np.log(arreglobase) / np.log(2))

print("Logaritmo en base 3")
print(np.log(arreglobase) / np.log(3))

print("Logaritmo en base 8")
print(np.log(arreglobase) / np.log(8))
```
